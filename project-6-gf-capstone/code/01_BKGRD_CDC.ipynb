```python
# Remove all Locations other than Alabama
new_df = new_df[new_df['LocationDesc'] == 'Alabama']
```

```python
### - not a good solution
# Fill missing values with the mode (most frequent value) for categorical variables
# for col in ['Education', 'Gender', 'Income', 'Age(years)']:
#     new_df[col] = new_df[col].fillna(new_df[col].mode()[0])

# Fill missing values with 'unknown' for categorical variables
for col in ['Education', 'Gender', 'Income', 'Age(years)', 'Data_Value', 'ClassID']:
    new_df[col] = new_df[col].fillna('unknown')
```

```python
new_df.info()
```

```python
# Remove rows where Data_Value is 'unknown'
new_df = new_df[new_df['Data_Value'] != 'unknown']

# Reset the index 
new_df = new_df.reset_index(drop=True)

# Check the new info
print(new_df.info())
```

```python
# Split dataframes into Class Topics

# Create dataframe with rows that include Physical Activity
df_PA = new_df[new_df['ClassID'] == 'PA']

# Create dataframe with rows that include Obesity/Weight Status
df_OWS = new_df[new_df['ClassID'] == 'OWS']

# Create dataframe with rows that include Fruits and Vegetables
df_FV = new_df[new_df['ClassID'] == 'FV']
```

```python
df_PA.shape
```

```python
df_OWS.shape
```

```python
df_FV.shape
```

```python
df_PA.info()
```

```python
# Convert Data_Value to float
df_PA['Data_Value'] = df_PA['Data_Value'].astype(float)

# Calculate the mean Data_Value for each Education category
education_means = df_PA.groupby('Education')['Data_Value'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=education_means.index, y=education_means.values)

# Customize the plot
plt.title('Average Data Value by Education Level (Physical Activity)', fontsize=16)
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Average Data Value', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Convert Data_Value to float, ignoring any errors
df_PA['Data_Value'] = pd.to_numeric(df_PA['Data_Value'], errors='coerce')

# Calculate the mean Data_Value for each Gender category
gender_means = df_PA.groupby('Gender')['Data_Value'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=gender_means.index, y=gender_means.values)

# Customize the plot
plt.title('Average Data Value by Gender (Physical Activity)', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Data Value', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Convert Data_Value to float, ignoring any errors
df_PA['Data_Value'] = pd.to_numeric(df_PA['Data_Value'], errors='coerce')

# Calculate the mean Data_Value for each Income category
income_means = df_PA.groupby('Income')['Data_Value'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=income_means.index, y=income_means.values)

# Customize the plot
plt.title('Average Data Value by Income (Physical Activity)', fontsize=16)
plt.xlabel('Income', fontsize=12)
plt.ylabel('Average Data Value', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Convert Data_Value to float, ignoring any errors
df_PA['Data_Value'] = pd.to_numeric(df_PA['Data_Value'], errors='coerce')

# Calculate the mean Data_Value for each Age category
age_means = df_PA.groupby('Age(years)')['Data_Value'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=age_means.index, y=age_means.values)

# Customize the plot
plt.title('Average Data Value by Age Group (Physical Activity)', fontsize=16)
plt.xlabel('Age Group (years)', fontsize=12)
plt.ylabel('Average Data Value', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
